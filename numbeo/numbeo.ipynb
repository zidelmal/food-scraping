```python
import requests
import pandas as pd
import json
from bs4 import BeautifulSoup as bs
```

```python
WILAYA_YASSIR = ["Algiers", "Annaba", "Ain-Defla", "Ain-Temouchent", "Batna", "Biskra", "Blida", "Bordj-Bou-Arreridj", "Bouira", "Boumerdes", "Bejaia", "Chlef", "Constantine", "Djelfa", "El-Oued", "El-Tarf", "Ghardaia", "Guelma", "Jijel", "Khenchela", "Laghouat", "M%27Sila", "Mascara", "Mila", "Mostaganem", "Medea", "Oran", "Ouargla", "Oum-El-Bouaghi", "Relizane", "Saida", "Sidi-Bel-Abbes", "Skikda", "Souk-Ahras", "Setif", "Tiaret", "Tipaza", "Tizi-Ouzou", "Tlemcen", "Tebessa"]
URL = 'https://www.numbeo.com/taxi-fare/in/'
```

```python
r = requests.get('https://www.numbeo.com/taxi-fare/in/Constantine-Algeria')
soup = bs(r.content,"html.parser")
```

```python
all_wilaya = soup.find_all('a', 'likenormal')
```

```python
with pd.ExcelWriter('numbeo.xlsx') as writer:
    for wilaya in all_wilaya:
        
        r = requests.get(wilaya.get('href'))
        soup = bs(r.content,"html.parser")
        data = pd.read_html(str(soup.find('div','standard_margin')))[0]
        data.to_excel(writer, sheet_name=wilaya.find_all('span')[1].get_text())
```

```python
for wilaya in WILAYA_YASSIR:
    try:
        if wilaya != 'Algiers':
            r = requests.get(URL + wilaya + '-Algeria')
        else:
            r = requests.get(URL + wilaya)
        soup = bs(r.content,"html.parser")
        data = pd.read_html(str(soup.find('div','standard_margin')))[0]
        data.to_csv(f"data/{wilaya}.csv", index=False)
    except:
        pass
```
