```python
import pandas as pd
```

```python
df = pd.read_csv('/home/yassir/Documents/work/yassir-scraping/uber-eats/uberEats.csv')
```

```python
def menu_(menu, row):
    url = []
    food_category = []
    prod_name = []
    description = []
    price = []
    for sec in menu['hasMenuSection']:
        for i in range(len(sec['hasMenuItem'])):
            try:
                a = sec['hasMenuItem'][i]['name']
                b = sec['hasMenuItem'][i]['description']
                c = sec['hasMenuItem'][i]['offers']['price']
                url.append(row['url'])
                food_category.append(sec['name'])
                prod_name.append(a)
                description.append(b)
                price.append(c)
            except:
                pass
    
    return {'url':url,
            'food_category': food_category,
            'prod_name': prod_name,
            'description' : description,
            'price': price}
```

```python
df_menu = pd.DataFrame()
for index, row in df.iterrows():
    try:
        menu = eval(row['menu'])
        df_menu = pd.concat([df_menu, pd.DataFrame(menu_(menu, row))
                    ]).reset_index(drop=True)
    except Exception as e:
        print(e)
        pass
    
```

```python
df = df.drop(columns='menu').drop_duplicates(subset='url').reset_index(drop=True)
```

```python
df_menu = df_menu.drop_duplicates().reset_index(drop=True)
```

```python
df.merge(df_menu, on='url', how='right').drop_duplicates().reset_index(drop=True).to_csv('UberEats_menu.csv', index=False)
```
